```python
import matplotlib.pyplot as plt
import pandas as pd

# Sample dataset of sales data
data = {
    'Product': ['Apples', 'Bananas', 'Cherries', 'Dates', 'Elderberries'],
    'Sales': [120, 80, 50, 90, 40]
}

# Convert to a DataFrame
df = pd.DataFrame(data)

```

```python
# 1. Histogram - Distribution of Sales
plt.figure(figsize=(10, 6))
plt.hist(df['Sales'], bins=5, color='skyblue', edgecolor='black')
plt.title('Histogram of Product Sales')
plt.xlabel('Sales')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()


```

```python
# 2. Bar Chart - Sales per Product
plt.figure(figsize=(10, 6))
plt.bar(df['Product'], df['Sales'], color='lightgreen')
plt.title('Bar Chart of Sales per Product')
plt.xlabel('Product')
plt.ylabel('Sales')
plt.show()

```

```python
# 3. Pie Chart - Sales Distribution by Product
plt.figure(figsize=(8, 8))
plt.pie(df['Sales'], labels=df['Product'], autopct='%1.1f%%', startangle=140, colors=['#ff9999','#66b3ff','#99ff99','#ffcc99','#c2c2f0'])
plt.title('Pie Chart of Sales Distribution')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```
