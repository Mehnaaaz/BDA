```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression, LogisticRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score, accuracy_score, confusion_matrix
from sklearn.datasets import make_classification

```

```python
# --- Linear Regression Example ---

# Sample dataset for Linear Regression
# Create a simple dataset
X_linear = np.random.rand(100, 1) * 10  # 100 random values between 0 and 10
y_linear = 2.5 * X_linear + 5 + np.random.randn(100, 1) * 2  # Linear relationship with some noise

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X_linear, y_linear, test_size=0.2, random_state=42)

# Linear Regression model
linear_model = LinearRegression()
linear_model.fit(X_train, y_train)

# Make predictions
y_pred_linear = linear_model.predict(X_test)

# Evaluate the model
mse = mean_squared_error(y_test, y_pred_linear)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_pred_linear)

print("Linear Regression Model Performance:")
print(f"RMSE: {rmse}")
print(f"R² Score: {r2}")

# Plot the results
plt.scatter(X_test, y_test, color='blue', label='True values')
plt.plot(X_test, y_pred_linear, color='red', label='Regression line')
plt.title('Linear Regression: True vs Predicted')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.show()


```

```python
from sklearn.datasets import make_classification

# Sample dataset for Logistic Regression
# Create a binary classification dataset with adjusted parameters
X_logistic, y_logistic = make_classification(n_samples=100, n_features=3, 
                                             n_informative=2, n_redundant=0, 
                                             n_repeated=0, n_classes=2, random_state=42)

# Train-test split
from sklearn.model_selection import train_test_split
X_train_logistic, X_test_logistic, y_train_logistic, y_test_logistic = train_test_split(X_logistic, y_logistic, test_size=0.2, random_state=42)

# Logistic Regression model
from sklearn.linear_model import LogisticRegression
logistic_model = LogisticRegression()
logistic_model.fit(X_train_logistic, y_train_logistic)

# Make predictions
y_pred_logistic = logistic_model.predict(X_test_logistic)

# Evaluate the model
from sklearn.metrics import accuracy_score, confusion_matrix
accuracy = accuracy_score(y_test_logistic, y_pred_logistic)
conf_matrix = confusion_matrix(y_test_logistic, y_pred_logistic)

print("\nLogistic Regression Model Performance:")
print(f"Accuracy: {accuracy}")
print(f"Confusion Matrix:\n{conf_matrix}")

# Plot the results
import matplotlib.pyplot as plt
plt.figure(figsize=(8, 6))
plt.scatter(X_test_logistic[:, 0], X_test_logistic[:, 1], c=y_pred_logistic, cmap=plt.cm.Paired)
plt.title('Logistic Regression: Predictions')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```
